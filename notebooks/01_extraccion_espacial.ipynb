# 01 · Extracción espacial

Identifica todas las referencias espaciales del corpus aplicando el catálogo del YAML. Produce el Excel con las referencias, su supuesto histórico, coordenadas aproximadas y etiquetas.

## 1. Cargar configuración y corpus

```python
CONFIG_PATH = 'configs/sin_remedio.yaml'
CORPUS_PATH = 'data/corpus.txt'

cfg = corpus.load_config(CONFIG_PATH)
parrafos = corpus.parse_from_config(CORPUS_PATH, cfg)
print(f"Párrafos parseados: {len(parrafos)}")
print(f"Primer párrafo: {parrafos[0]['parrafo_id']} ({parrafos[0]['capitulo']})")
```

## 2. Cargar el catálogo de lugares

```python
cat_path = cfg['lugares_catalogo_path']
with open(cat_path, encoding='utf-8') as f:
    catalogo = json.load(f)
print(f"Catálogo: {len(catalogo)} entradas")
```

## 3. Extraer referencias espaciales

```python
df = spatial_extraction.extract(parrafos, catalogo)
print(f"Referencias extraídas: {len(df)}")
df.head(10)
```

## 4. Resúmenes

```python
res_lugar = spatial_extraction.summary_by_place(df)
res_lugar.head(15)
```

```python
res_cap = spatial_extraction.summary_by_chapter(df)
res_cap
```

## 5. Exportar Excel

```python
sheets = {
    'Extracción espacial': {
        'df': df,
        'title': 'Extracción espacial completa',
        'subtitle': cfg['proyecto']['nombre']
    },
    'Por lugar': {'df': res_lugar, 'title': 'Resumen por lugar'},
    'Por capítulo': {'df': res_cap, 'title': 'Referencias por capítulo'},
}

out = exporters.to_excel(sheets, 'outputs/01_extraccion_espacial.xlsx')
print(f"✓ Guardado en {out}")
```
