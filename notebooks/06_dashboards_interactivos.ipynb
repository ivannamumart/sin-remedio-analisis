## 1. Mapa afectivo interactivo

````python
# Cargar el GeoJSON producido en el notebook 02
places = json.loads(Path('outputs/lugares_emocional.geojson').read_text(encoding='utf-8'))

# Convertir features de GeoJSON a la estructura que espera interactive.affective_map_html
places_flat = []
for f in places['features']:
    p = f['properties'].copy()
    p['lon'], p['lat'] = f['geometry']['coordinates']
    places_flat.append(p)

interactive.affective_map_html(places_flat, 'outputs/mapa_afectivo.html')
print('✓ outputs/mapa_afectivo.html')
````
